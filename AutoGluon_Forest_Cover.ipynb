```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("/content/covtype.csv")
```

```python
data.head()
```

```python
train, test = train_test_split(data,random_state=42, test_size = 0.3)
y_test = test['Cover_Type']
test_nolab = test.drop(['Cover_Type'],axis=1)
```

```python
from autogluon.tabular import TabularPredictor
```

```python
predictor = TabularPredictor(label='Cover_Type').fit(train)
```

```python
pred_probs = predictor.predict_proba(test_nolab)
pred_probs.head(5)
```

```python
# Leaderboard for the trained model
predictor.leaderboard(test, silent=True)
```
